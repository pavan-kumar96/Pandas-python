### Understanding the pivot table

```python
gap_data = pd.read_csv('https://raw.githubusercontent.com/edyoda/data-science-complete-tutorial/master/Data/gapminder-FiveYearData.csv')
```

```python
gap_data.pivot_table(index = "country", columns = "year", values = "lifeExp")
```

```python
### find out the most improved country in terms of life expententy
gap_data.describe()
```

```python
def f(r):
    return(float(r.lifeExp.max())- float(r.lifeExp.min()))
gap_data["improve_life"] = gap_data.lifeExp.map(f)
```
