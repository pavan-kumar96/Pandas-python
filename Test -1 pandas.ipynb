###
3. Find where more people consume alcohol in rural or urban

```python
rural = hc_df.apply(lambda x: True if x["Place(location where the patient lives)"] == "rural" else False, axis = 1)
# print(rural.value_counts())
urban = hc_df.apply(lambda x: True if x["Place(location where the patient lives)"] == "urban" else False, axis = 1)

count_rural = len(rural[rural == True].index)
print(count_rural)

count_urban =  len(urban[urban == True].index)
print(count_urban)
if count_rural > count_urban:
    print("Rural people will consume more alcohol")
     
```
