```python
###2.Convert average_montly_hours into days & hours
```

```python
def f(r):
    return (str(r//24)+ "days"+ str(r%24)+"hours")
hr_data["Work_hours"] = hr_data.average_montly_hours.map(f)
hr_data
```

```python
def f(r):
    return (str(r.average_montly_hours//24)+ "days"+ str(r.average_montly_hours%24)+"hours")
hr_data["Work_hours"] = hr_data.apply(f, axis = 1)
hr_data
```

```python

t_data = pd.read_csv('https://raw.githubusercontent.com/edyoda/data-science-complete-tutorial/master/Data/titanic-train.csv.txt', index_col='PassengerId')
t_data
```

```python
###3.new column with values KID,ADULT & OLD based on age
```

```python
def f(r):
    if r.Age < 20:
        return "Child"
    elif r.Age>20 and r.Age<40:
        return "Adult"
    elif r.Age> 40:
        return "OLD"
t_data["Age_type"] = t_data.apply(f, axis =1)
```
