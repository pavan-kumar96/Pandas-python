```python
import pandas as pd 

import xml.etree.ElementTree as etree

tree = etree.parse("student.xml")

root = tree.getroot()
columns = ["id","name","salary"]
dataframe = pd.DataFrame(columns = columns)
for node in root: 

    name = node.find("id")

    mail = node.find("name").text if node is not None else None

    age = node.find("salary").text if node is not None else None

    dataframe = dataframe.append(pd.Series([id, mail, age], index = columns), ignore_index = True)
dataframe
```

```python
import xml.etree.ElementTree as et
import pandas as pd
tree = et.parse("student.xml")
root = tree.getroot()
columns =  ["id","name","salary"]
df = pd.DataFrame(columns = columns)
for node in root:
    id = node.find("id").text
    name = node.find("name").text
    salary = node.find("salary").text
    df = df.append(pd.Series([id, name, salary], index = columns),ignore_index = True)
df
```
