```python
def f(g):
    if (g== "sales ") | (g == "marketing"):
        return "S&M"
hr_data["new_data"] = hr_data.deparment.map(f)
```

```python
hr_data.sample(10)
```

```python
def f(r):
    if (r.satisfaction_level>0.82) & (r.salary == "low")& (r.left == 0):
        return "medium"
    else:
        return r.salary
hr_data["New_salary"] = hr_data.apply(f, axis =1)
```
